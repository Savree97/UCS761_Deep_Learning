```python
import numpy as np
np.random.seed(42)
```

```python
#creating the dataset
X=np.linspace(0,10,50).reshape(-1,1)
print(X)
```

```python
noise=np.random.normal(0,0.2,size=(50,1))
y=np.log(X+1)+noise

# Target grows non-linearly (log shape) with small noise.
# A straight line cannot fit both fast initial growth and later slow growth.
```

```python
hidden_units=3

# We have chosen 3 hidden units,more than 1 allows bending.
# Too many increases instability and complexity.
```

```python
#initalizing parameters

W1=np.random.uniform(-1,1,size=(1,hidden_units))
b1=np.zeros((1,hidden_units))
W2=np.random.uniform(-1,1,size=(hidden_units,1))
b2=np.zeros((1,1))

# W1 controls how input influences hidden layer (shape bending).
# b1 shifts hidden layer outputs.
# W2 combines hidden units to produce output.
# b2 shifts final prediction.
```

```python
#defining activation function
def activation(z):
    return np.maximum(0,z)

def activation_slope(z):
    return (z>0).astype(float)
```

```python
learning_rate=0.01
epochs=1000

for epoch in range(epochs):

    # Forward pass
    z1=X@W1+b1
    h=activation(z1)
    y_hat=h@W2+b2

    # Error and loss
    error=y_hat-y
    loss=np.mean(error**2)

    # Backward pass
    dL_dy=2*error/len(X)

    # Output layer gradients
    dL_dW2=h.T@dL_dy
    dL_db2=np.sum(dL_dy,axis=0,keepdims=True)

    # Hidden layer gradients
    dL_dh=dL_dy@W2.T
    dL_dz1=dL_dh*activation_slope(z1)

    dL_dW1=X.T@dL_dz1
    dL_db1=np.sum(dL_dz1,axis=0,keepdims=True)

    # Update rule (move opposite to gradient)
    W1-=learning_rate*dL_dW1
    b1-=learning_rate*dL_db1
    W2-=learning_rate*dL_dW2
    b2-=learning_rate*dL_db2

    if epoch%100==0:
        print("Epoch:",epoch,"Loss:",loss)
```
