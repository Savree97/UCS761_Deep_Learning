```python
import random
import math
```

```python
def step(z):
    return 1 if z >= 0 else 0

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(y):
    return y * (1 - y)
```

```python
def train_single_layer(X, Y, lr=0.1, epochs=20):
    w1 = random.uniform(-1, 1)
    w2 = random.uniform(-1, 1)
    b  = random.uniform(-1, 1)

    for epoch in range(epochs):
        total_error = 0

        for (x1, x2), y in zip(X, Y):
            z = w1*x1 + w2*x2 + b
            y_hat = step(z)
            error = y - y_hat

            w1 += lr * error * x1
            w2 += lr * error * x2
            b  += lr * error

            total_error += abs(error)

        print(f"Epoch {epoch+1} | Total Error = {total_error}")

    return w1, w2, b
```

```python
def test_single_layer(name, X, Y):
    print(f"\n{name} Gate Training (Single-Layer)")
    w1, w2, b = train_single_layer(X, Y)

    print("Final Results:")
    for (x1, x2), y in zip(X, Y):
        z = w1*x1 + w2*x2 + b
        print(f"{x1} {x2} -> {step(z)} (expected {y})")
```

```python
def train_mlp(X, Y, lr=0.5, epochs=5000):
    w1 = [[random.uniform(-1,1), random.uniform(-1,1)],
          [random.uniform(-1,1), random.uniform(-1,1)]]
    b1 = [random.uniform(-1,1), random.uniform(-1,1)]

    w2 = [random.uniform(-1,1), random.uniform(-1,1)]
    b2 = random.uniform(-1,1)

    for epoch in range(epochs):
        total_error = 0

        for (x1, x2), y in zip(X, Y):
            h1 = sigmoid(w1[0][0]*x1 + w1[0][1]*x2 + b1[0])
            h2 = sigmoid(w1[1][0]*x1 + w1[1][1]*x2 + b1[1])
            out = sigmoid(w2[0]*h1 + w2[1]*h2 + b2)

            error = y - out
            total_error += error**2

            d_out = error * sigmoid_derivative(out)
            d_h1 = d_out * w2[0] * sigmoid_derivative(h1)
            d_h2 = d_out * w2[1] * sigmoid_derivative(h2)

            w2[0] += lr * d_out * h1
            w2[1] += lr * d_out * h2
            b2    += lr * d_out

            w1[0][0] += lr * d_h1 * x1
            w1[0][1] += lr * d_h1 * x2
            b1[0]    += lr * d_h1

            w1[1][0] += lr * d_h2 * x1
            w1[1][1] += lr * d_h2 * x2
            b1[1]    += lr * d_h2

        if epoch % 500 == 0:
            print(f"Epoch {epoch} | Error = {total_error:.4f}")

    return w1, b1, w2, b2
```

```python
def predict_mlp(x1, x2, w1, b1, w2, b2):
    h1 = sigmoid(w1[0][0]*x1 + w1[0][1]*x2 + b1[0])
    h2 = sigmoid(w1[1][0]*x1 + w1[1][1]*x2 + b1[1])
    out = sigmoid(w2[0]*h1 + w2[1]*h2 + b2)
    return 1 if out >= 0.5 else 0
```

```python
X = [(0,0), (0,1), (1,0), (1,1)]

AND  = [0,0,0,1]
OR   = [0,1,1,1]
NAND = [1,1,1,0]
NOR  = [1,0,0,0]
XOR  = [0,1,1,0]
XNOR = [1,0,0,1]
```

```python
test_single_layer("AND", X, AND)
test_single_layer("OR", X, OR)
test_single_layer("NAND", X, NAND)
test_single_layer("NOR", X, NOR)

print("\nXOR Gate Training (Multi-Layer)")
w1, b1, w2, b2 = train_mlp(X, XOR)
for x1, x2 in X:
    print(x1, x2, "->", predict_mlp(x1, x2, w1, b1, w2, b2))

print("\nXNOR Gate Training (Multi-Layer)")
w1, b1, w2, b2 = train_mlp(X, XNOR)
for x1, x2 in X:
    print(x1, x2, "->", predict_mlp(x1, x2, w1, b1, w2, b2))
```
